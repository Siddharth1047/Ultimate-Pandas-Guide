##### Sorting the data:

```python
df.sort_values('Lifespan')
# By default 'sort_values()' sorts the data in ascending format
```

```python
# Sorting in descending order
df.sort_values('Lifespan', ascending = False)
```

##### Sort using two columns

```python
df.sort_values(['Lifespan', 'Ratings'], ascending = [True, False])
# This will sort the data in the following order:- 
# First it'll sort on the basis of 'Lifespan' column in ascending format
# Then it'll sort on the basis of 'Ratings' column in descending format
```

##### What is the highest lifespan of an animal ?

```python
df['Lifespan'].max()
```

##### Which animals possess a lifespan of 20+ ?

```python
df_filter = df[df['Lifespan'] > 20]
names = df_filter['Name']
```
